```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon.csv")
df.head()
```

```python
# Başlıkları Türkçeleştirme
df.columns = [
    'urun_id',           # Product ID
    'urun_ismi',         # Name of the Product
    'kategori',          # Category of the Product
    'indirimli_fiyat',   # Discounted Price of the Product
    'asil_fiyat',        # Actual Price of the Product
    'indirim_yuzdesi',   # Percentage of Discount for the Product
    'derecelendirme',    # Rating of the Product
    'derecelendirme_sayisi',  # Number of people who voted for the Amazon rating
    'urun_bilgisi',      # Description about the Product
    'kullanici_id',      # ID of the user who wrote review for the Product
    'kullanici_adi',     # Name of the user who wrote review for the Product
    'inceleme_id',       # ID of the user review
    'inceleme_basligi',  # Short review
    'inceleme_icerigi',  # Long review
    'resim_baglanti',    # Image Link of the Product
    'urun_baglanti'      # Official Website Link of the Product
]

df.to_csv("amazon_turkce.csv", index=False)
```

```python
nan_counts = df.isna().sum()
print("Sütunlardaki NaN değer sayıları:{}".format(nan_counts))
# NaN değerlerinin görselleştirilmesi
plt.figure(figsize=(12, 6))
sns.barplot(x=nan_counts.index, y=nan_counts.values, palette='viridis')
plt.xlabel('Sütun Adı')
plt.ylabel('NaN Değer Sayısı')
plt.title('Veri Setindeki NaN (Eksik) Değerlerin Sayısı')
plt.xticks(rotation=45)
plt.show()
```

```python
# Kategorileri ayırarak her bir kategori için ayrı bir satır oluşturma
df['kategori'] = df['kategori'].str.split('&|\\|')  # Hem '&' hem de '|' işaretlerini ayırır
df = df.explode('kategori')  # Kategorileri satırlara dökme

# Sonuçları kontrol etme
print(df[['urun_ismi', 'kategori']].head(n=30))
```

```python
# ₹ işaretini kaldırma ve sayısal formata çevirme
df['indirimli_fiyat'] = df['indirimli_fiyat'].replace({'₹': ''}, regex=True)
df['asil_fiyat'] = df['asil_fiyat'].replace({'₹': ''}, regex=True)

# Sayısal formata dönüştürme
df['indirimli_fiyat'] = pd.to_numeric(df['indirimli_fiyat'], errors='coerce')
df['asil_fiyat'] = pd.to_numeric(df['asil_fiyat'], errors='coerce')

# Hatalı veya eksik veri olan satırları temizleme
df = df.dropna(subset=['indirim_yuzdesi', 'asil_fiyat', 'indirimli_fiyat'])  # NaN olan satırları kaldırma

# Sonuçları kontrol etme
print(df[['urun_ismi', 'indirim_yuzdesi', 'asil_fiyat', 'indirimli_fiyat']].head(n = 20))
```

```python
# İndirim yüzdesine göre azalan sırayla sıralama
en_fazla_indirimli_urunler = df.sort_values(by='indirim_yuzdesi', ascending=False)

# En yüksek indirimli 10 ürünü seçme
en_fazla_indirimli_urunler_top_10 = en_fazla_indirimli_urunler[['urun_ismi', 'indirim_yuzdesi', 'kategori']].head(20)

# Sonuçları gösterme
print("En Yüksek İndirimli Ürünler ve Kategorileri:")
print(en_fazla_indirimli_urunler_top_10)
```

ASIL FİYAT VE İNDİRİMLİ FİYAT ARASINDAKİ FARKIN EN BÜYÜK OLDUĞU ÜRÜNLERİ GETİRİR.

```python
# Yüzde işaretlerini kaldırarak sayısal formata dönüştürme
df['indirim_yuzdesi'] = df['indirim_yuzdesi'].replace({'%': ''}, regex=True)
df['indirim_yuzdesi'] = pd.to_numeric(df['indirim_yuzdesi'], errors='coerce')

# Aynı işlemi 'asil_fiyat' ve 'indirimli_fiyat' için de yapalım (₹ işaretini kaldırma)
df['asil_fiyat'] = df['asil_fiyat'].replace({'₹': '', ',': ''}, regex=True)
df['asil_fiyat'] = pd.to_numeric(df['asil_fiyat'], errors='coerce')

df['indirimli_fiyat'] = df['indirimli_fiyat'].replace({'₹': '', ',': ''}, regex=True)
df['indirimli_fiyat'] = pd.to_numeric(df['indirimli_fiyat'], errors='coerce')


# En fazla indirimli 10 ürünü alırken tüm sütunları dahil edelim
top_10_indirimli = df[['urun_ismi', 'indirim_yuzdesi', 'kategori', 'asil_fiyat', 'indirimli_fiyat']].sort_values(by='indirim_yuzdesi', ascending=False).head(100)


```

```python
# Tekrar eden ürünleri kaldırmak için drop_duplicates kullanıyoruz
top_10_indirimli_unique = top_10_indirimli.drop_duplicates(subset='urun_ismi')

# Fiyat farkını hesapla
top_10_indirimli_unique['fiyat_farki'] = top_10_indirimli_unique['asil_fiyat'] - top_10_indirimli_unique['indirimli_fiyat']

# Fiyat farkına göre sıralama (en büyük farktan en küçük farka)
top_10_indirimli_sorted = top_10_indirimli_unique.sort_values(by='fiyat_farki', ascending=False)

# Farklı görselleştirme: Asıl fiyat ve indirimli fiyat arasındaki farkı göstermek için yatay bar plot
plt.figure(figsize=(12, 10))

# Asıl fiyat ve indirimli fiyat arasındaki farkı göstermek için yatay bar plot
sns.barplot(
    x='asil_fiyat', y='urun_ismi', data=top_10_indirimli_sorted, color='lightblue', label='Asıl Fiyat'
)

sns.barplot(
    x='indirimli_fiyat', y='urun_ismi', data=top_10_indirimli_sorted, color='orange', label='İndirimli Fiyat'
)

# Başlık ve etiketler
plt.xlabel('Fiyat (₹)')
plt.ylabel('Ürün İsmi')
plt.title('Asıl Fiyat ve İndirimli Fiyat Karşılaştırması (Fiyat Farkına Göre Sıralı)')

# Efsane (Legend)
plt.legend(title="Fiyat Türü")

# Fiyat farklarını, asıl fiyat ve indirimli fiyat ile birlikte hesaplayıp konsola yazdırma
print("Ürünler ve Fiyat Farkları (En Büyük Farktan Küçüğe):\n")
for i, row in top_10_indirimli_sorted.iterrows():
    # Fiyat farkını hesapla
    fiyat_farki = row['fiyat_farki']
    
    # Fiyat farklarını ve fiyatları yazdır
    print(f"Ürün: {row['urun_ismi']}")
    print(f"  Asıl Fiyat: ₹{row['asil_fiyat']:.2f}")
    print(f"  İndirimli Fiyat: ₹{row['indirimli_fiyat']:.2f}")
    print(f"  Fiyat Farkı: ₹{fiyat_farki:.2f}\n")

# Grafik düzenini optimize et
plt.tight_layout()

# Görselleştirmeyi göster
plt.show()
```

```python
# Tekrar eden ürünleri kaldırıyoruz
top_10_indirimli_unique = top_10_indirimli.drop_duplicates(subset='urun_ismi')

# Fiyat farkını hesapla
top_10_indirimli_unique['fiyat_farki'] = top_10_indirimli_unique['asil_fiyat'] - top_10_indirimli_unique['indirimli_fiyat']

# Fiyat farkına göre sıralama (en büyük farktan en küçük farka)
top_10_indirimli_sorted = top_10_indirimli_unique.sort_values(by='fiyat_farki', ascending=False)

# Fiyat farkları ve diğer bilgilerle yeni bir DataFrame oluşturma
fiyat_farklari_tablo = top_10_indirimli_sorted[['urun_ismi', 'asil_fiyat', 'indirimli_fiyat', 'fiyat_farki', 'kategori']]

# Tabloyu yazdırma
print("Fiyat Farkları ve Ürün Kategorileri Tablosu:")
print(fiyat_farklari_tablo)

# Eğer isterseniz tabloyu CSV olarak da kaydedebilirsiniz
# fiyat_farklari_tablo.to_csv('fiyat_farklari_tablo.csv', index=False)
```

BU ÜRÜNLERİN KATEGORİLERİNİ DE SAYALIM VE EN YÜKSEK FİYAT FARKI  EN ÇOK HANGİ KATEGORİLERDE GÖZÜKÜYOR BİLELİM.

```python
# Tekrar eden ürünleri kaldırıyoruz
top_10_indirimli_unique = top_10_indirimli.drop_duplicates(subset='urun_ismi')

# Fiyat farkını hesapla
top_10_indirimli_unique['fiyat_farki'] = top_10_indirimli_unique['asil_fiyat'] - top_10_indirimli_unique['indirimli_fiyat']

# Fiyat farkına göre sıralama (en büyük farktan en küçük farka)
top_10_indirimli_sorted = top_10_indirimli_unique.sort_values(by='fiyat_farki', ascending=False)

# Fiyat farkları ve diğer bilgilerle yeni bir DataFrame oluşturma
fiyat_farklari_tablo = top_10_indirimli_sorted[['urun_ismi', 'asil_fiyat', 'indirimli_fiyat', 'fiyat_farki', 'kategori']]

# Kategorileri ayırıyoruz (bir üründe birden fazla kategori olabilir)
kategori_listesi = top_10_indirimli_sorted['kategori'].str.split('&|\|', expand=True).stack().reset_index(drop=True)

# Kategorileri sayıyoruz
kategori_sayilari = kategori_listesi.value_counts().reset_index()
kategori_sayilari.columns = ['Kategori', 'Adet']

# Kategorilerin sayısını yazdırıyoruz
print("Kategori Sayıları:")
print(kategori_sayilari)


```

```python
# En yüksek fiyat farkına sahip ürünleri seçiyoruz
top_10_indirimli_sorted = top_10_indirimli_sorted.sort_values(by='fiyat_farki', ascending=False)

# Kategorilerdeki adetleri sayma
# Kategorileri & veya | ile ayırarak birden fazla kategori olabilen ürünleri tek tek sayacağız
category_counts = top_10_indirimli_sorted['kategori'].str.split('&|\|', expand=True).stack().value_counts()

# Grafik için boyut ayarlamaları
plt.figure(figsize=(12, 8))

# Bar plot (Kategoriler ve adetleri)
sns.barplot(x=category_counts.values, y=category_counts.index, palette='coolwarm')

# Başlık ve etiketler
plt.title('En Yüksek Fiyat Farkına Sahip Ürünlerin Yer Aldığı Kategorilerin Adetleri', fontsize=14)
plt.xlabel('Kategori Adedi', fontsize=12)
plt.ylabel('Kategori', fontsize=12)

# Göster
plt.show()
```
